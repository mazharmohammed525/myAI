# what is List
- A list is a built-in data type used to store multiple items in a single variable.
- A list is an ordered, changeable (mutable) collection of values.
## Key features of a list
- Ordered → items have a fixed position (index starts at 0)
- Mutable → you can change, add, or remove items
- Allows duplicates → same value can appear more than once
- Can store different data types → numbers, strings, even other lists

----------------------------------------------------------------------

# SET
- A set is a built-in data type used to store a collection of unique, unordered element

### Key features of set:

- unordered --> Items don't have a fixed position( no indexing like lists)
- No duplicates --> each value appears only once
- Mutable --> you can add or remove items after creation

## Add and update methods of set 
- Add method is used to add single value to the set 
- Update method is used to add multiple values to the set. 

### Discard, Remove, Clear methods of set
- Discard and Remove methods are used to delete the values from set.
- If the element is present on set, both methods will work fine.
- If the element is not present then discard will return none whereas 
- remove method will give a key error.
- Clear method is used to make an empty set
- Del is used to delete the set 

### Set operators : union, intersection, difference, symmetric difference
- Union will return unique values from both the sets
- Intersection will return the common values from both the sets.
- Difference will return values from first set which are not present on the 
second set.
- Symmetric difference will return unique values except common values 
from both sets.

# Dictionaries
- A dictionaries is a built in data types used to store data in key-value paurs

## Key features of dictionaries
- Key-Value strucuture --> each item has a key and its value
- Keys must be unique -  no duplicates allowed
- Mutable --> you can change, add, or remove items
- Ordered(python 3.7+) --> Items keep insertion order

# Numpy
- Numpy (short for Numerical Pyhton) is a powerful open-source library n python used for numerical comuting.Its one of the core building bloks for data science, machine learning, generative AI and Agentic AI..etc.

### what numpy does?
- multi-dimensional array
- Fast arrays: A special data structure called an  ndarray(n-dimensional array) that stores numbers efficiently.
- Mathematical operations: Funcations for performaing calculations like addition, mulitplication,trignometry,stats  -- often much fater than plain python
- Linear algebra tools: matrix operations,eigenvalues, dot products etc. 
- Random number generation:used for simulations and ML

### Why its important
- Speed: Numpy is implemented in optimised C code, so it runs much faster than a regular python loops
- Convenience: You can perform operations on entire arrays without writing loops.
- Foundation for other libraries. Many popular tools like Pandas, Matplotlib, and scikit-learn are built on top of Numpy
- Handling numerical data inside agents
- working with emdeddings(vectors)
- Doing similarity calculations(cosine similarty etc)
- supporting backend computations when needed

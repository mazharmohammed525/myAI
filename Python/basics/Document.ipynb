## Boolean Variables
- Boolean Variables are variables that can have only two possible values: True or False
- A boolean variables answers a yes/no question

# what is List
- A list is a built-in data type used to store multiple items in a single variable.
- A list is an ordered, changeable (mutable) collection of values.
## Key features of a list
- Ordered → items have a fixed position (index starts at 0)
- Mutable → you can change, add, or remove items
- Allows duplicates → same value can appear more than once
- Can store different data types → numbers, strings, even other lists

# SET
- A set is a built-in data type used to store a collection of unique, unordered element

### Key features of set:

- unordered --> Items don't have a fixed position( no indexing like lists)
- No duplicates --> each value appears only once
- Mutable --> you can add or remove items after creation

# Dictionaries
- A dictionaries is a built in data types used to store data in key-value paurs

## Key features of dictionaries
- Key-Value strucuture --> each item has a key and its value
- Keys must be unique -  no duplicates allowed
- Mutable --> you can change, add, or remove items
- Ordered(python 3.7+) --> Items keep insertion order

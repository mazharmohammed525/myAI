# what is List
- A list is a built-in data type used to store multiple items in a single variable.
- A list is an ordered, changeable (mutable) collection of values.
## Key features of a list
- Ordered → items have a fixed position (index starts at 0)
- Mutable → you can change, add, or remove items
- Allows duplicates → same value can appear more than once
- Can store different data types → numbers, strings, even other lists

```python
type([])
```

```python
listexample = [1,2,3,4,5]
print(listexample)
print(type(listexample))
```

# SET
- A set is a built-in data type used to store a collection of unique, unordered element

### Key features of set:

- unordered --> Items don't have a fixed position( no indexing like lists)
- No duplicates --> each value appears only once
- Mutable --> you can add or remove items after creation

```python
myset =set() # it creates an empty set. It does not take any parameters.  It returns an empty set.
print(myset)
print(type(set))
```

```python
myset = {1,1,2,3,4}
```

```python
print(myset)
```

```python
myset = {'ML','AI','DS','ML'}
print(myset)
```

```python
print(myset)
```

```python
myset.add("math")  # it adds an element to the set. it takes one parameter: the element to add.It does not return any value
print(myset)
```

```python
myset[0]  # it returns a TypeError becuase set do not support indexing
```

```python
myset1 = {'ML','AI','DS'}
myset2 = {'ML','AI','DS','math'}
```

```python
myset2.difference(myset1) # it retuns a set that contains difference between two sets. 
```

```python
myset1.intersection(myset2)  # this return the common elements from both sets
```

```python
myset1.union(myset2)  # this will return all the elements from both the sets and it avoids duplications
```

```python
myset1.intersection_update(myset2)  # it will remove the elements from original set that are not present in the specified set
print(myset1)
print(myset2)
```

```python
myset2.intersection_update(myset1)
print(myset1)
print(myset2)
```

```python
print(myset1)
print(myset2)
```

# Dictionaries
- A dictionaries is a built in data types used to store data in key-value paurs

## Key features of dictionaries
- Key-Value strucuture --> each item has a key and its value
- Keys must be unique -  no duplicates allowed
- Mutable --> you can change, add, or remove items
- Ordered(python 3.7+) --> Items keep insertion order

# Numpy
- Numpy (short for Numerical Pyhton) is a powerful open-source library n python used for numerical comuting.Its one of the core building bloks for data science, machine learning, generative AI and Agentic AI..etc.

### what numpy does?
- multi-dimensional array
- Fast arrays: A special data structure called an  ndarray(n-dimensional array) that stores numbers efficiently.
- Mathematical operations: Funcations for performaing calculations like addition, mulitplication,trignometry,stats  -- often much fater than plain python
- Linear algebra tools: matrix operations,eigenvalues, dot products etc. 
- Random number generation:used for simulations and ML

### Why its important
- Speed: Numpy is implemented in optimised C code, so it runs much faster than a regular python loops
- Convenience: You can perform operations on entire arrays without writing loops.

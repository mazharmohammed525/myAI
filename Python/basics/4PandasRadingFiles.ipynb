```python
# creating a dataframe with the help of numpy
df= pd.DataFrame(np.arange(0,20).reshape(5,4))
df
```

```python
# giving names to Rows and columns
df= pd.DataFrame(np.arange(0,20).reshape(5,4),index=['Row1','Row2','Row3','Row4','ROw5'],columns=['Col1','Col2','Col3','Col4'])
print(df)
```

```python
# Save the data frame as csv file, the file will be add in the folder
df.to_csv("myDataFrame.csv")
```

```python
# to access the elements of the dataframe
# there are two options
# 1) .loc  --> this can handle strings, Rows, columns
# 2). iloc --> this can handle only integers.(position of index)

print(df['Col1'])
```

```python
print(df.loc['Row1'])
```

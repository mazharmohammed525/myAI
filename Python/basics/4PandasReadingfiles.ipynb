# To read the file from RAM/memory for better performance, below are the two ways

# Use cases of StringIO
- Temporary text storage
- csv processing
- Capturing printed output
- Testing file operations without creating files

### Example: Reading csv from memory

# Use cases of BytesIO
- Image processing
- Audio/video handling
- File uploads/downloads
- Working with API's
- Machine learning model files

### Example: Image Handling

# Difference between String IO and Bytes IO

| Feature | StringIO | BytesIO|
|--- | --- | --- |
| Data Type | Text('str') | Binary ('bytes')|
| works like | Text file | Binary file |
| Input Example | "Hello" | b"Hello" |
| common Use | CSV/Text | Images/PDFs |

- stringIO --> notebook for text
- BytesIO --> storage box for binary files
- Both are virtual files in RAM

```python
# to import StringIO,BytesIO, we need to import IO library
from io import StringIO,BytesIO
```

```python
data = ('col1,col2,col3\n'
        'a,b,1\n'
        'c,d,2\n'
        'e,f,3')
```

```python
type(data)
```

```python
# to read the file in pandas
pd.read_csv(StringIO(data))
```

```python
# get specific columns
df = pd.read_csv(StringIO(data),usecols=['col1','col2'])
print(df)
```

```python
df.to_csv('test2.csv')
```

```python
#Specifying columns data types
data = ('a,b,c,d\n'
        '1,2,3,4\n'
        '5,6,7,8\n'
        '9,10,11,12')
```

```python
print(data)
print(type(data))
```

```python
df = pd.read_csv(StringIO(data))
print(df)
```

```python
# to make the columns as string 
df = pd.read_csv(StringIO(data),dtype='object')
print(df)
```

```python
df['a']
```

```python
# I want to make column b as Int and c as string

df = pd.read_csv(StringIO(data),dtype={'b':int,'c':float})
print(df)
```

```python
# to check datatypes for all rows in df
df.dtypes
```

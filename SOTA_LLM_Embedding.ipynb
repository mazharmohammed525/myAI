```python
from sentence_transformers import SentenceTransformer
```

```python
sentence_embedding_model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
# one word embedding and assigning to a variable
mytext= 'mazhar'
```

```python
my_embedding = sentence_embedding_model.encode(mytext)
```

```python
my_embedding
```

```python
# to check the length of Embedding
len(my_embedding)
```

```python
# to check the shape of matrix
my_embedding.shape
```

```python
# one sentence embedding

my_sentence = "This is an example sentence for embedding."
my_sentence_embedding = sentence_embedding_model.encode(my_sentence)
```
